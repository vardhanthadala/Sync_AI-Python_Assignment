### 26. Star Repetition – 3
**Approach:** Use logic to keep boundaries intact while filling the center with stars in a reusable function.

```python
def problem_26():
    s = input()
    if len(s) >= 2:
        print(s[0] + "*" * (len(s) - 2) + s[-1])
    else:
        print(s)

problem_26()
```
