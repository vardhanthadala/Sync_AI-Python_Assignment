### 1. Sum of Two Numbers
**Approach:** Define a function that reads two integers from input, calculates their sum, and prints the result.

```python
def sum_of_two_numbers():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print(a + b)

sum_of_two_numbers()
```

### 2. Division of Two Numbers
**Approach:** Use a function to read two integers and perform float division to get the result.

```python
def division_of_two_numbers():
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))
    print(num1 / num2)

division_of_two_numbers()
```

### 3. Area of a Rectangle
**Approach:** Calculate the area by multiplying the length and breadth obtained from user input.

```python
def area_of_rectangle():
    length = int(input("Enter length: "))
    breadth = int(input("Enter breadth: "))
    print(length * breadth)

area_of_rectangle()
```

### 4. Perimeter of Rectangle
**Approach:** Use the formula 2 * (length + breadth) within a function to calculate the perimeter.

```python
def perimeter_of_rectangle():
    length = int(input("Enter length: "))
    breadth = int(input("Enter breadth: "))
    print(2 * (length + breadth))

perimeter_of_rectangle()
```

### 5. Integer Division

```python
def integer_division():
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    print(a // b)

integer_division()
```

### 12. Sum of Digits
**Approach:** Convert the number to a string, iterate through characters, convert back to integers, and sum them.

```python
def sum_of_digits():
    num = input("Enter number: ")
    total = sum(int(digit) for digit in num)
    print(total)

sum_of_digits()
```
