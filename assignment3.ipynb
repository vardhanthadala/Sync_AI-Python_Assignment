# SyncAi Intern Assignment 3 – Boolean Comparisons

### 1. Greater Than 70

```python
num = int(input("Enter number: "))
print(num > 70)
```

### 2. Compare Two Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(a > b)
```

### 3. Negative Number Check

```python
num = int(input("Enter number: "))
print(num < 0)
```

### 4. Word Comparison

```python
word1 = input("Enter first word: ")
word2 = input("Enter second word: ")
print(word1 == word2)
```

### 5. Sunday Check

```python
day = int(input("Enter day number: "))
print(day == 7)
```

### 6. Numbers Not Equal

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print(num1 != num2)
```

### 7. Greater Than or Equal Comparison

```python
a = float(input("Enter A: "))
b = float(input("Enter B: "))
condition = a >= b
print(f"A >= B is {condition}")
```

### 8. Less Than or Equal Check

```python
a = int(input("Enter first number: "))
b = float(input("Enter second number: "))
print(a <= b)
```

### 9. Consecutive Numbers

```python
a = int(input("Enter A: "))
b = int(input("Enter B: "))
print(b == a + 1)
```

### 10. First and Last Letter Check

```python
word = input("Enter word: ")
print(word[0] != word[-1])
```

### 11. Sum of Digits Check

```python
num_str = input("Enter two-digit number: ")
digit_sum = int(num_str[0]) + int(num_str[1])
print(digit_sum > 7)
```

### 12. Word Length Check

```python
word = input("Enter word: ")
print(len(word) > 7)
```
