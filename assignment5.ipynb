### 14. Electricity Bill

```python
units = int(input())
bill = 0
if units <= 50:
    bill = units * 2
elif units <= 150:
    bill = (50 * 2) + (units - 50) * 3
elif units <= 250:
    bill = (50 * 2) + (100 * 3) + (units - 150) * 5
else:
    bill = (50 * 2) + (100 * 3) + (100 * 5) + (units - 250) * 8

total_bill = bill + (bill * 0.20)
print(float(total_bill))
```
